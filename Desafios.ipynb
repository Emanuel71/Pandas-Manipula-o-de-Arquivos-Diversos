## Desafio 05

Você é responsável por criar um banco de dados local de clientes para uma instituição financeira. Temos o arquivo CSV com os dados de clientes. Realizar três atualizações no banco de dados:


```python
# Importando as bibliotecas 
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect, text

# Criando o motor do banco de dados
engine = create_engine('sqlite:///:memory:')

#Importando os dados csv
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
dadosclientes = pd.read_csv(url)

#Colocando DataFrame no banco de dados
dadosclientes.to_sql('dadosClientes', engine, index=False)
```

```python
# Verificando se a tabela está no banco de dados criado
inspecao = inspect(engine)
print(inspecao.get_table_names())
```

```python
pd.read_sql_table('dadosClientes', engine).head(3)
```

```python
# Atualizar o registro do cliente de ID 6840104 que teve o rendimento anual alterado para 300000.
query1 = 'UPDATE dadosClientes SET Rendimento_anual = 300000.0 WHERE ID_Cliente = 6840104'

with engine.connect() as conn:
    conn.execute(text(query1))
    conn.commit()

pd.read_sql_table('dadosClientes', engine)
```

```python
# Excluir o registro do cliente de ID 5008809, pois essa pessoa não possui mais conta na instituição financeira.
query2 = 'DELETE FROM dadosClientes WHERE ID_Cliente = 5008809 '
with engine.connect() as conn:
    conn.execute(text(query2))
    conn.commit()

pd.read_sql_table('dadosClientes', engine)
```

```python
# Criar um novo registro de cliente seguindo as especificações
query3 = 'INSERT INTO dadosClientes(ID_Cliente,Idade,Grau_escolaridade,Estado_civil,Tamanho_familia,Categoria_de_renda,Ocupacao,Anos_empregado,Rendimento_anual,Tem_carro,Moradia) VALUES(6850985,33,"Doutorado", "Solteiro", 1, "Empregado", "TI", 2, 290000.0, 0, "Casa/apartamento próprio")'
with engine.connect() as conn:
    conn.execute(text(query3))
    conn.commit()

pd.read_sql_table('dadosClientes', engine).tail(5)
```
