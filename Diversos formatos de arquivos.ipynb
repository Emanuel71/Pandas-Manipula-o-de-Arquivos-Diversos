## Aula 05 - Trabalhando com Dados

O time de dados de uma Instituição Financeira possui um arquivo CSV que contém os dados de todos os clientes cadastrados. Porém, o que eles querem é pegar esse arquivo, colocar em um banco de dados local e a partir dele realizar consultas SQL. Assim será possível analisar dados, obter insights e tomar decisões estratégicas.

```python
#Importando bibliotecas 
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect, text

# Criando o banco de dados 
engine = create_engine('sqlite:///:memory:')

# Importando os dados.csv
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
dados = pd.read_csv(url)

# Colocando esse df no banco de dados
dados.to_sql('clientes', engine, index=False)
```

```python
# Inspecionando, para ver se incluirmos a tabela no banco de dados
inspecionar = inspect(engine)
print(inspecionar.get_table_names())
```

```python
# Iremos realizar consular sql 
# Precisamos apenas dos clientes que tenha categorias de renda 'empregadas'
query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado"'

#Visualizando a seleção
empregados = pd.read_sql(query, engine)
empregados
```

```python
# Salvando a consulta
empregados.to_sql('empregados', con=engine, index=False)
```

```python
# Visualizando uma tabela interia 
pd.read_sql_table('empregados', engine)
```

```python
# Visualizando determinas colunas 
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual']).head()
```

```python
# Para deletar um registro na tabela clientes, agora utilizamos a seguinte estrutura:
query1 = 'DELETE FROM clientes WHERE ID_Cliente = 5008804'
with engine.connect() as conn:
    result = conn.execute(text(query1))
    conn.commit()
    
pd.read_sql_table('clientes', engine).head(9)
```

```python
# De maneira similar, para atualizar informações de um registro, como o grau de escolaridade de um cliente
query2 = 'UPDATE clientes SET Grau_escolaridade = "Ensino Superior" WHERE ID_Cliente = 5008808'
with engine.connect() as conn:
    result = conn.execute(text(query2))
    conn.commit()

pd.read_sql_table('clientes', engine).head(9)
```
